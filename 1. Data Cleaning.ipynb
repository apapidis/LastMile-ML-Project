```python
import json
import pandas as pd
import os
```

```python
list_data = ['route_data', 'package_data', 'travel_times', 'actual_sequences']
path = 'almrrc2021-data-training/model_build_inputs/'

with open(path + 'route_data.json', 'r') as f:
    route_data = json.load(f)

with open(path + 'package_data.json', 'r') as f:
    package_data = json.load(f)

with open(path + 'travel_times.json', 'r') as f:
    travel_times = json.load(f)

with open(path + 'actual_sequences.json', 'r') as f:
    actual_sequences = json.load(f)
```

```python
len(route_data), len(package_data), len(travel_times), len(actual_sequences)
```

```python
first_route = list(route_data.keys())[0]
first_route in package_data and first_route in travel_times and first_route in actual_sequences
```

```python
routes_list = []
route_data[first_route].keys()
```

```python
routes_list = []

for route_id, route_info in route_data.items():
    routes_list.append({
        'route_id': route_id,
        'station_code': route_info['station_code'],
        'date': route_info['date_YYYY_MM_DD'],
        'departure_time_utc': route_info['departure_time_utc'],
        'executor_capacity_cm3': route_info['executor_capacity_cm3'],
        'route_score': route_info['route_score']
    })

routes_df = pd.DataFrame(routes_list)
routes_df.head(4)
```

```python
routes_df.shape
```

```python
stops_list = []
for route_id, route_info in route_data.items():
    stops = route_info['stops']
    for stop_id, stop_info in stops.items():
        
        stops_list.append({
            'route_id': route_id,
            'stop_id': stop_id,
            'lat': stop_info["lat"],
            'lng': stop_info["lng"],
            "type": stop_info["type"],
            'zone_id': stop_info["zone_id"]})

stops_df = pd.DataFrame(stops_list)
stops_df.head(4)
```

```python
stops_df.shape
```

```python
package_data[first_route]

packages_list = []

for route_id, route_info in package_data.items():
    for stop_id, stop_info in route_info.items():
        for package_id, package_info in stop_info.items():

            packages_list.append({
                'route_id': route_id,
                'stop_id': stop_id,
                'package_id': package_id,
                'scan status': package_info['scan_status'],
                'start_time_utc':package_info['time_window']['start_time_utc'],
                'end_time_utc':package_info['time_window']['end_time_utc'],
                'planned_service_time_seconds': package_info['planned_service_time_seconds'],
                'depth_cm': package_info['dimensions']['depth_cm'],
                'height_cm': package_info['dimensions']['height_cm'],
                'width_cm': package_info['dimensions']['width_cm']})

packages_df = pd.DataFrame(packages_list)
packages_df.head(4)      
```

```python
actual_sequences

sequences_list = []

for route_id, route_info in actual_sequences.items():
    actual = route_info['actual']
    for stop_id, stop_order in actual.items():
        sequences_list.append({
            'route_id': route_id,
            'stop_id': stop_id,
            'stop_order': stop_order})

sequences_df = pd.DataFrame(sequences_list)

sequences_df.head(4)
```

```python
sequences_df
```

```python
sequences_sorted = sequences_df.sort_values(['route_id', 'stop_order']).copy()
sequences_df.sort_values(['route_id', 'stop_order']).copy()

sequences_sorted['next_stop_id'] = (
    sequences_sorted.groupby('route_id')['stop_id'].shift(-1))

legs_df = (sequences_sorted.dropna(subset=['next_stop_id'])
    .rename(columns={'stop_id': 'from_stop', 'next_stop_id': 'to_stop'})
    [['route_id', 'from_stop', 'to_stop', 'stop_order']]
    .copy())

legs_df.head()
```

```python
def get_travel_time(row):
    route_id = row['route_id']
    from_stop = row['from_stop']
    to_stop = row['to_stop']
    return travel_times[route_id][from_stop][to_stop]

legs_df['travel_time_seconds'] = legs_df.apply(get_travel_time, axis=1)
```

```python
legs_df.head()
```

```python
route_travel_duration = (legs_df.groupby('route_id')['travel_time_seconds']
    .sum().reset_index(name='total_travel_duration'))

route_service_time = (packages_df.groupby('route_id')['planned_service_time_seconds']
                      .sum().reset_index(name='total_service_time_duration'))

route_duration_df = route_travel_duration.merge(route_service_time,
                                             on='route_id',
                                             how='left')

route_duration_df['ERD_seconds'] = (route_duration_df['total_travel_duration']
                                    + route_duration_df['total_service_time_duration'])

route_duration_df['ERD_hours'] = route_duration_df["ERD_seconds"] / 3600

route_duration_df.sort_values('ERD_hours', ascending = False).head(4)
```

```python
route_df = routes_df.merge(route_duration_df,
                           on='route_id', how='left')
route_df.head(4)
```

```python
stops_df.head(4)

route_num_stops= (stops_df.groupby('route_id')['stop_id']
                  .count().reset_index(name='num_stops'))
route_df = route_df.merge(route_num_stops, on='route_id', how='left')
route_df.head(4)
```

```python
packages_df.head(2)

route_packages = (packages_df.groupby('route_id')['package_id']
                  .count().reset_index(name='total_packages'))

route_df = route_df.merge(route_packages, on='route_id', how='left')
route_df.head(4)
```

```python
route_df['packages_per_stop'] = (route_df['total_packages'] / route_df['num_stops'])

route_df['avg_service_time'] = (
    route_df['total_service_time_duration'] / route_df['total_packages'])

route_df.head(4)
```

```python
route_df['capacity_m3'] = route_df['executor_capacity_cm3']/1000000
route_df.drop(columns='executor_capacity_cm3', inplace=True)
route_df.head()
```

```python
import os
os.makedirs("data", exist_ok=True)
route_df.to_csv("data/route_dataset.csv", index=False)
```
